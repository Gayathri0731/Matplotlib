6. Grid( )

```python
import matplotlib.pyplot as plt
```

```python
x = [22,34,45,54]
y = [32,45,57,42]
```

```python
plt.grid(axis='x')
```

```python
plt.plot(x,y)
plt.grid(axis='x') #vertical
```

```python
plt.plot(x,y)
plt.grid(axis='y') #horizontal
```

Specifing color and style of the line

```python
plt.grid(color='red',linestyle='--', linewidth=0.8)
```
