5. Piechart--> pie( )

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Labels =['A','B','C','D']
sizes = [25,20,30,25]
```

```python
plt.pie(sizes,labels=Labels)
```

```python
y = np.array([32,28,15,25])
```

```python
plt.pie(y)
```

Explode : 
Make one wedge to stands out

```python
z = np.array([38,22,15,25])
label = ['A','B','C','D']
```

```python
exp = [0, 0, 0.2, 0]
```

```python
plt.pie(z, labels=label, explode=exp)
```

Shadow

```python
plt.pie(z, labels=label, explode=exp, shadow=True)
```

Colors

```python
x_1 = np.array([45,35,20])

labels = ['A','B','C']

color = ['#00cccc','Pink','#d11aff']
```

```python
plt.pie(x_1, labels=labels,colors=color)
```

Legend --> plt . legend( )

```python
y_1 = np.array([30,20,25,25])

lab = ['Apple','Orange','Banana','Cherry']
```

```python
plt.pie(y_1, labels=lab)
plt.legend(y_1)
```

```python
plt.pie(y_1, labels=lab)
plt.legend(y_1,labels=lab)
```

```python
plt.pie(y_1, labels=lab)
plt.legend(y_1, labels=lab, title ="Fruits")
```
