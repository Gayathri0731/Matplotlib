2. Scatter( )

Syntax : plt.scatter(x,y)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = (1,3,4,9)
y = (56,67,69,80)
```

```python
plt.scatter(x,y)
```

We can also choose colors

```python
## choose color through name
plt.scatter(x,y,color='red')
```

```python
## choosing color through the color value

plt.scatter(x,y,color='#cc33ff')
```

```python
x1 = (20,30,40)
y1 = (90,80,70)
```

```python
clrs = np.array(['red','green','teal'])
plt.scatter(x1,y1,c=clrs)    ## c == color
```
