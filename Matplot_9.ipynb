Label & Title

```python
import matplotlib.pyplot as plt
```

```python
x = [15,20,26,31]
y = [32,35,25,40]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature- *C')

plt.plot(x,y)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature- *C')

plt.plot(x,y)

plt.title('Weather')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature- *C')

plt.plot(x,y)

plt.title('Weather',loc='left')
```

```python
plt.plot(x,y)

plt.xlabel('Days')
plt.ylabel('Temperature- *C')

plt.title('Weather', fontdict={
    'family': 'serif',
    'color': 'teal',
    'weight': 'bold',
    'size': 16,
    'style': 'italic',
    'backgroundcolor': 'pink'
})
```
